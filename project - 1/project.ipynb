```python
def extract_eps(txt):
    check=False
    data=""

    for i in txt:
        if i=='(':
            check = True
            continue
        if i==')':
            break
        if check:
            data+=i

    return data
            
                
        
```

```python
df["Episode"]=df['Title'].apply(extract_eps)
df.head() 
#Extracting the Episode in differnt coloumn
```

```python
df["Episode"]=df["Episode"].str.replace("eps", "")  #removing the string
```

```python
df["Episode"]=df["Episode"].astype(int)
df
#Converting into int
```

```python
def extraction_time(txt):
    check=False
    data=""
    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1,i+20):
                data+=txt[j]

            return data
            #Extracting the time
```

```python
df['Time']=df["Title"].apply(extraction_time)
df.head()
```

```python
start = pd.to_datetime(df["Time"].str.split(" - ").str[0], format="%b %Y")
end   = pd.to_datetime(df["Time"].str.split(" - ").str[1], format="%b %Y")

df["Months"] = (end.dt.year - start.dt.year) * 12 + (end.dt.month - start.dt.month)
df.head()
#Extracting the month
```
